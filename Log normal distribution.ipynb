```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
sns.set(rc={'figure.figsize': (11.7, 8.27)})
g = sns.barplot(x = 'income($)', y = 'count', data =df)
g.set_xticklabels(g.get_xticklabels(), rotation = 45, horizontalalignment = 'right');
```
